# Create and Test Model

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# South
# Initialize model
model_S = RandomForestRegressor(n_estimators=300, random_state=42)
# Train model
model_S.fit(X_train_S, y_train_S)
```

```python
# East
# Initialize model
model_E = RandomForestRegressor(n_estimators=300, random_state=42)
# Train model
model_E.fit(X_train_E, y_train_E)
```

```python
# West
# Initialize model
model_W = RandomForestRegressor(n_estimators=300, random_state=42)
# Train model
model_W.fit(X_train_W, y_train_W)
```

```python
# Predict with model and see accuracy of Training set
# South
y_train_pred_S = model_S.predict(X_train_S)
mse_train_S = mean_squared_error(y_train_S, y_train_pred_S)
r2_train_S = r2_score(y_train_S, y_train_pred_S)

# East
y_train_pred_E = model_E.predict(X_train_E)
mse_train_E = mean_squared_error(y_train_E, y_train_pred_E)
r2_train_E = r2_score(y_train_E, y_train_pred_E)

# West
y_train_pred_W = model_W.predict(X_train_W)
mse_train_W = mean_squared_error(y_train_W, y_train_pred_W)
r2_train_W = r2_score(y_train_W, y_train_pred_W)
```

```python
print('Training MSE South:', mse_train_S)
print('Training R^2 South:', r2_train_S)
print()
print('Training MSE East:', mse_train_E)
print('Training R^2 East:', r2_train_E)
print()
print('Training MSE West:', mse_train_W)
print('Training R^2 West:', r2_train_W)
```

```python
# Test prediction and accuracy check
# South
y_test_pred_S = model_S.predict(X_test_S)
mse_test_S = mean_squared_error(y_test_S, y_test_pred_S)
r2_test_S = r2_score(y_test_S, y_test_pred_S)

print('Test MSE South:', mse_test_S)
print('Test R^2 South:', r2_test_S)
print()
# East
y_test_pred_E = model_E.predict(X_test_E)
mse_test_E = mean_squared_error(y_test_E, y_test_pred_E)
r2_test_E = r2_score(y_test_E, y_test_pred_E)

print('Test MSE East:', mse_test_E)
print('Test R^2 East:', r2_test_E)
print()
# West
y_test_pred_W = model_W.predict(X_test_W)
mse_test_W = mean_squared_error(y_test_W, y_test_pred_W)
r2_test_W = r2_score(y_test_W, y_test_pred_W)

print('Test MSE West:', mse_test_W)
print('Test R^2 West:', r2_test_W)
```

```python
# Visualize random forest model's results
import matplotlib.pyplot as plt

# Scatter plot of actual vs. predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test_S, y_test_pred_S, alpha=0.5)
plt.plot([y_test_S.min(), y_test_S.max()], [y_test_S.min(), y_test_S.max()], 'k--', lw=2)  # Diagonal line
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```
