```python
# Fixes test shape to get rid of the NaN values
#Before dropping NaN values from y_test_S, remove corresponding rows from X_test_S
#X_test_S = X_test_S.loc[y_test_S.index]  # Use the index of y_test_S to align with X_test_S
X_test_S = X_test_S.dropna().reset_index(drop=True)
print('X_test South shape',X_test_S.shape)

# Now drop NaN values from y_test_S
X_test_E = X_test_E.dropna().reset_index(drop=True)
print('X_test East shape', X_test_E.shape)

# Reset index after dropping NaN values (if y_test_S was a pandas Series or DataFrame)
X_test_W = X_test_W.reset_index(drop=True)
print('X_test West shape',X_test_W.shape)
```

# Create and Test Model

```python
# Import model
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.impute import SimpleImputer

# Fit the imputer to your training data and transform both training and testing data
imputer = SimpleImputer(strategy='mean')

# South
X_train_S = imputer.fit_transform(X_train_S)
X_test_S = imputer.transform(X_test_S)

# East
X_train_E = imputer.fit_transform(X_train_E)
X_test_E = imputer.transform(X_test_E)

# West
X_train_W = imputer.fit_transform(X_train_W)
X_test_W = imputer.transform(X_test_W)
```

```python
# South
# Implement model (just trying with South right now)
model_S = SVR(kernel='linear')
#model.fit(X_train_S_imputed, y_train_S_imputed)
model_S.fit(X_train_S, y_train_S)
```

```python
# East
# Implement model
model_E = SVR(kernel='linear')

# Test model
model_E.fit(X_train_E, y_train_E)
```

```python
# West
# Implement model
model_W = SVR(kernel='linear')

# Test model
model_W.fit(X_train_W, y_train_W)
```

```python
# Predict data
# South
y_train_pred_S = model_S.predict(X_train_S)
mse_train_S = mean_squared_error(y_train_S, y_train_pred_S)
r2_train_S = r2_score(y_train_S, y_train_pred_S)

# East
y_train_pred_E = model_E.predict(X_train_E)
mse_train_E = mean_squared_error(y_train_E, y_train_pred_E)
r2_train_E = r2_score(y_train_E, y_train_pred_E)

# West
y_train_pred_W = model_W.predict(X_train_W)
mse_train_W = mean_squared_error(y_train_W, y_train_pred_W)
r2_train_W = r2_score(y_train_W, y_train_pred_W)
```

```python
print('Training MSE South:', mse_train_S)
print('Training R^2 South:', r2_train_S)
print()
print('Training MSE East:', mse_train_E)
print('Training R^2 East:', r2_train_E)
print()
print('Training MSE West:', mse_train_W)
print('Training R^2 West:', r2_train_W)
```

```python
# South
# Predictions
y_test_pred_S = model_S.predict(X_test_S)
print(y_test_pred_S.shape)

y_test_S_reset = y_test_S.reset_index(drop=True) # resets dataframe

# Calculate mse and r2
mse_test_S = mean_squared_error(y_test_S.reset_index(drop=True), y_test_pred_S)
r2_test_S = r2_score(y_test_S, y_test_pred_S)

print('Test MSE South:', mse_test_S)
print('Test R^2 South:', r2_test_S)
```

```python
# East
# Predictions
y_test_pred_E = model_E.predict(X_test_E)
print(y_test_pred_E.shape)

y_test_E_reset = y_test_E.reset_index(drop=True) # resets dataframe

# Calculate mse and r2
mse_test_E = mean_squared_error(y_test_E.reset_index(drop=True), y_test_pred_E)
r2_test_E = r2_score(y_test_E, y_test_pred_E)

print('Test MSE East:', mse_test_E)
print('Test R^2 East:', r2_test_E)
```

```python
# West
# Predictions
y_test_pred_W = model_W.predict(X_test_W)
print(y_test_pred_W.shape)

y_test_W_reset = y_test_W.reset_index(drop=True) # resets dataframe

# Calculate mse and r2
mse_test_W = mean_squared_error(y_test_W.reset_index(drop=True), y_test_pred_W)
r2_test_W = r2_score(y_test_W, y_test_pred_W)

print('Test MSE West:', mse_test_W)
print('Test R^2 West:', r2_test_W)
```
